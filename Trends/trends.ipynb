```python
from pytrends.request import TrendReq
import pandas as pd
import matplotlib.pyplot as plt
```

```python
pytrends = TrendReq(hl='en-US', tz=360)
```

```python
keyword = ['Climate change']
```

```python
import time

# Add a delay of 10 seconds
time.sleep(10)

# Make the request
pytrends.build_payload(keyword, cat=0, timeframe='2020-01-01 2025-01-01', geo='', gprop='')
```

```python
data = pytrends.interest_over_time()
```

```python
# Plot the interest over time data
data.plot(y=keyword, figsize=(10, 5), title='Interest over time')
plt.xlabel('Date')
plt.ylabel('Interest')
plt.show()
```
