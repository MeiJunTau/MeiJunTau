```python
from pytrends.request import TrendReq

# Initialize pytrends request
pytrends = TrendReq(hl='en-GB', tz=360)

# Build payload for the keyword(s) you are interested in
kw_list = ["Electric Car"]
pytrends.build_payload(kw_list, cat=0, timeframe='today 12-m', geo='GB', gprop='')

# Get interest over time
interest_over_time_df = pytrends.interest_over_time()
```

```python
import matplotlib.pyplot as plt

# Plot the interest over time
interest_over_time_df['Electric Car'].plot(figsize=(10, 6))
plt.title('Interest Over Time for Electric Car')
plt.xlabel('Date')
plt.ylabel('Interest')
plt.grid(True)
plt.show()
```

```python
# Initialize pytrends request
pytrends = TrendReq(hl='en-GB', tz=360)

# Build payload for the keyword(s) you are interested in
kw_list = ["Electric Car","Heat Pump","Solar Panel"]
pytrends.build_payload(kw_list, cat=0, timeframe='today 5-y', geo='GB', gprop='')

# Get interest over time
interest_over_time_df = pytrends.interest_over_time()
```

```python
# Plot the interest over time for all keywords
interest_over_time_df[kw_list].plot(figsize=(12, 8))
plt.title('Interest Over Time for Keywords')
plt.xlabel('Date')
plt.ylabel('Interest')
plt.legend(title='Keywords')
plt.grid(True)
plt.show()
```
