```python
import pandas as pd
from pytrends.request import TrendReq
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.animation as animation
```

```python
import time
```

```python
pytrends = TrendReq(hl='en-US', tz=360)
```

```python
keywords = ['bitcoin', 'AI','Climate Change','Quantum Computing','Electric Cars']
pytrends.build_payload(keywords, cat=0, timeframe='2024-01-01 2024-12-31', geo='GB')
```

```python
# Add a delay of 10 seconds before making the request
time.sleep(10)
df = pytrends.interest_over_time()
```

```python
df = df.drop(columns='isPartial')
```

```python
# Plot the interest over time data
df.plot(y=keywords, figsize=(10, 5), title='Interest over time')
plt.xlabel('Date')
plt.ylabel('Interest')
plt.show()
```
